1. Explain the differences between supervised, semi-surpervised, weakly-supervised and unsupervised learning.

**Answer:**

Supervised learning is a type of machine learning algorithm that learns from labeled data. The machine learns the relationship between input features and corresponding outputs. 

Semi- supervised learning is a type of machine learning that uses both labeled and unlabeled data to train a model for classification or regression tasks. It is useful when there is a lot of unlabeled data available compared to the labeled data.

Weakly supervised learning is a type of machine learning trained with partial or noisy labels, which are less precise than fully labeled data.

Unsupervised learning is a type of learning where the algorithm is given data without explicit labels. The system tries to find patterns, relationships, or structures in the data without being told what to look for.

7. Write the objective function that can be used to determine the regression model parameters. How is this objective function will be used to find model parameters?

**Answer:** 

An objective function that can be used for regression is the mean squared error (MSE), which is the average of the squared differences between the predicted and true values. The MSE can be written as:
MSE(Œ∏)= $1/N \sum_{i=1}^{N}(f(x_i‚à£Œ∏)-y_i )^2$
where Œ∏ are the model parameters, $f(x_i‚à£Œ∏)$ is the prediction for the i-th input, and $y_i$ is the true value for the i-th input.

To find the optimal parameters, we can use optimization techniques such as gradient descent that iteratively update the parameters based on the gradient of the objective function. This approach involves minimizing the cost function.Gradient descent  updates the parameters as:
$\theta \leftarrow \theta - \alpha \frac{\partial MSE(\theta)}{\partial \theta}$
where Œ± is the learning rate, and $\frac{\partial MSE(\theta)}{\partial \theta}$ is the gradient of the MSE with respect to the parameters. The gradient tells us the direction and magnitude of the steepest ascent of the objective function, so by moving in the opposite direction, we can reduce the objective function value.

**Part III**
       
Gradient descent to find the optimal fit.
1. Initialize learning rate and epoch, try to explain your reasons for the values chosen;

**Question: How did you choose your lr and epoch number?**

**Answer:**
I set the learning rate to be 0.8 in order to make the optimisation procedure more stable and effective, and epoch number to be 1000 to make sure the model will converge.



```python
# gradient descent to find the optimal fit
def gradient_descent(X,y,theta,learning_rate,epochs):
    cost_history = []
    n = len(y)
    
    for epoch in range(epochs):
        predictions = theta[0] + theta[1] * X
        error = predictions - y
        
        gradient_0 = (1/n) * np.sum(error)
        gradient_1 = (1/n) * np.sum(error*X) 
        
        theta[0] = theta[0] - learning_rate * gradient_0
        theta[1] = theta[1] - learning_rate * gradient_1
        
        cost = (1/(2*n)) * np.sum(error**2)
        cost_history.append(cost)
    return theta, cost_history 
```

```python
epochs = 1000
learning_rate = 0.8
final_theta, cost_history = gradient_descent(slr_train_X, slr_train_y, initial_theta, learning_rate, epochs)
print("\n")
print("The final theta is: ", final_theta)
print("\n")
print("The final cost: ", cost_history[-1])
```

# 2. Multiple Linear Regression

```python
# Do not edit the codes in this cell
# select 8 specific columns as attributes

cols = [0,2,3,4,5,6,8,9]

# Fetch 8 attributes for multiple linear regression
mlr_train_X = train_X[:, cols]
mlr_train_y = train_y[:]

mlr_test_X = test_X[:, cols]
mlr_test_y = test_y[:]
```

**Part I**

Initialize theta and construct cost function for the upcoming gradient decent.
1. Initialize theta, print it out;
2. Construct Mean Square Error loss function;
3. Calcualte the cost using initial theta, X and y, and print the result out.


```python
# initialize theta
initial_theta_mlr = np.random.rand(len(cols))
print("\n")
print("The initial theta for mlr is:", initial_theta_mlr)
```

```python
# define cost function

def mse_loss(X,y,theta):
    n = len(y)
    predictions = np.dot(X, theta)
    error = predictions - y
    mse = (1/(2*n)) * np.sum(error**2)
    return mse

# calculate cost by calling the function
cost_mlr = mse_loss(mlr_train_X, mlr_train_y, initial_theta_mlr)

print("\n")
print("The cost for mlr is: ", cost_mlr)
```

**Part II**

Gradient descent to find the optimal fit.
1. Initialize learning rate and epoch, try to explain your reasons for the values chosen;
2. Construct gradient descent function, which updates the theta and meanwhile records all the history cost;
3. Call the function for the optimal fit. Print out the final theta and final cost.


**Question: How did you choose your lr and epoch number?**

**Answer:**

Chose epochs = 1003 in order to prevent the model from overfitting and to see if the model performance is improving and
learning_rate_mlr = 0.8 to help the model to converge


```python
# gradient descent to find the optimal fit

def grad_descent_mlr(X,y,theta,learning_rate,epochs):
    cost_history = []
    
    for epoch in range(epochs):
        predictions = np.dot(X, theta)      
       # y_reshaped = y.reshape(-1,1)    
        errors = predictions - y      
        gradients = (1 / len(y)) * np.dot(X.T, errors) 
        theta = theta - learning_rate * gradients  
        cost = mse_loss(X, y, theta)
        cost_history.append(cost)
    return theta, cost_history   
```

```python
epochs = 1003
learning_rate_mlr = 0.8 

final_theta_mlr, cost_history_mlr = grad_descent_mlr(mlr_train_X, mlr_train_y, initial_theta_mlr, learning_rate_mlr, epochs)
print("\n")
print("Final theta for MLR is: ", final_theta_mlr)
print("\n")
print("The final cost for mlr: ", cost_history_mlr[-1])
```
